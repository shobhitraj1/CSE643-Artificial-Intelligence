```python
def categorize_price(price):
    if price < 10000000:
        return 'Low'
    elif 10000000 <= price < 100000000:
        return 'Medium'
    elif 100000000 <= price < 200000000:
        return 'High'
    else:
        return 'Very High'
    
train_category = train.copy()

train_category['Price_Category'] = train_category['Price'].apply(categorize_price)

category_distribution = train_category['Price_Category'].value_counts()
print(category_distribution)

plt.figure(figsize=(8, 6))
category_distribution.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red'], edgecolor='black')
plt.title('Distribution of Properties Across Price Categories', fontsize=14)
plt.xlabel('Price Category', fontsize=12)
plt.ylabel('Number of Properties', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

imbalance_info = category_distribution / category_distribution.sum() * 100
print("\nPercentage Distribution Across Categories:\n", imbalance_info)
```

### Building Decision Tree Model

<h3 align='center'>Task 1</h3>

```python
from sklearn.tree import DecisionTreeRegressor

X_train = train.drop('Price', axis=1)
y_train = train['Price']

dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))
plot_tree(dt_regressor, feature_names=X_train.columns, filled=True, fontsize=10)
plt.show()

depth = dt_regressor.get_depth()
n_leaves = dt_regressor.get_n_leaves()
print(f"Depth of the tree: {depth}")
print(f"Number of leaves: {n_leaves}")
```

<h3 align='center'>Task 2</h3>

```python
feature_importances = dt_regressor.feature_importances_
features = X_train.columns
feature_importances_df = pd.DataFrame({'Features': features, 'Importance': feature_importances})
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Features', data=feature_importances_df)
plt.title("Feature Importances", fontsize=14)
plt.xlabel("Importance", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.show()

print(feature_importances_df)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [None, 10, 15, 20, 25],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_features': [None, 'sqrt', 'log2']
}

dt = DecisionTreeRegressor(random_state=42)

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=10, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
print(grid_search.best_score_)
print(grid_search.best_estimator_)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
X_test = test.drop('Price', axis=1)
y_test = test['Price']
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_train_pred = dt_regressor.predict(X_train)
y_train_pred_best = best_model.predict(X_train)

rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2_train = r2_score(y_train, y_train_pred)

rmse_train_best = np.sqrt(mean_squared_error(y_train, y_train_pred_best))
r2_train_best = r2_score(y_train, y_train_pred_best)

print("Default Model Performance on Train Data:")
print(f"RMSE: {rmse_train}")
print(f"R2 Score: {r2_train}")

print("\nBest Model Performance on Train Data:")
print(f"RMSE: {rmse_train_best}")
print(f"R2 Score: {r2_train_best}")

y_test_pred = dt_regressor.predict(X_test)
y_test_pred_best = best_model.predict(X_test)

rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_test = r2_score(y_test, y_test_pred)

rmse_test_best = np.sqrt(mean_squared_error(y_test, y_test_pred_best))
r2_test_best = r2_score(y_test, y_test_pred_best)

print("\nDefault Model Performance on Test Data:")
print(f"RMSE: {rmse_test}")
print(f"R2 Score: {r2_test}")

print("\nBest Model Performance on Test Data:")
print(f"RMSE: {rmse_test_best}")
print(f"R2 Score: {r2_test_best}")
```

```python
feature_importances_best = best_model.feature_importances_
features = X_train.columns

feature_importances_df_best = pd.DataFrame({'Features': features, 'Importance': feature_importances_best})
feature_importances_df_best = feature_importances_df_best.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Features', data=feature_importances_df_best)
plt.title("Feature Importances of the Best Model", fontsize=14)
plt.xlabel("Importance", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.show()

print(feature_importances_df_best)
```

<h3 align='center'>Task 3</h3>

```python
path = best_model.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

ccp_alphas
```

```python
path = best_model.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

ccp_alphas

pruned_dt_regressors = []

cnt = 0

for ccp_alpha in ccp_alphas:
    cnt += 1
    pruned_regressor = DecisionTreeRegressor(random_state=42, ccp_alpha=ccp_alpha, **best_params)
    pruned_regressor.fit(X_train, y_train)
    pruned_dt_regressors.append(pruned_regressor)
    # if cnt % 100 == 0:
    #     print(f"Completed {cnt} iterations")

train_scores = [pruned_regressor.score(X_train, y_train) for pruned_regressor in pruned_dt_regressors]
test_scores = [pruned_regressor.score(X_test, y_test) for pruned_regressor in pruned_dt_regressors]

plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_scores, marker='o', label='Train', drawstyle='steps-post')
plt.plot(ccp_alphas, test_scores, marker='o', label='Test', drawstyle='steps-post')
plt.xlabel("CCP Alpha", fontsize=12)
plt.ylabel("R2 Score", fontsize=12)
plt.title("R2 Score vs. CCP Alpha", fontsize=14)
plt.legend()
plt.show()
```

```python
best_alpha = ccp_alphas[np.argmax(test_scores)]
print(f"Best Alpha: {best_alpha}")
```

```python
best_alpha = ccp_alphas[np.argmax(test_scores)]
best_alpha

pruned_dt_regressor = DecisionTreeRegressor(random_state=42, ccp_alpha=best_alpha, **best_params)
pruned_dt_regressor.fit(X_train, y_train)

y_train_pred_pruned = pruned_dt_regressor.predict(X_train)
y_test_pred_pruned = pruned_dt_regressor.predict(X_test)

rmse_train_pruned = np.sqrt(mean_squared_error(y_train, y_train_pred_pruned))
r2_train_pruned = r2_score(y_train, y_train_pred_pruned)

rmse_test_pruned = np.sqrt(mean_squared_error(y_test, y_test_pred_pruned))
r2_test_pruned = r2_score(y_test, y_test_pred_pruned)

print("Pruned Model Performance on Train Data:")
print(f"RMSE: {rmse_train_pruned}")
print(f"R2 Score: {r2_train_pruned}")

print("\nPruned Model Performance on Test Data:")
print(f"RMSE: {rmse_test_pruned}")
print(f"R2 Score: {r2_test_pruned}")

plt.figure(figsize=(20, 10))
plot_tree(pruned_dt_regressor, feature_names=X_train.columns, filled=True, fontsize=10)
plt.show()
```

```python
feature_importances_pruned = pruned_dt_regressor.feature_importances_
features = X_train.columns

feature_importances_df_pruned = pd.DataFrame({'Features': features, 'Importance': feature_importances_pruned})
feature_importances_df_pruned = feature_importances_df_pruned.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Features', data=feature_importances_df_pruned)
plt.title("Feature Importances of Pruned Model", fontsize=14)
plt.xlabel("Importance", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.show()

print(feature_importances_df_pruned)
```

```python
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(random_state=42)
X_adasyn, y_adasyn = adasyn.fit_resample(X, y)

print("Original Class Distribution:\n", y.value_counts())
print("\nADASYN Class Distribution:\n", y_adasyn.value_counts())

X_adasyn_features = X_adasyn.drop('Price', axis=1)
y_adasyn_target = X_adasyn['Price']

dt_regressor_adasyn = DecisionTreeRegressor(random_state=42)
dt_regressor_adasyn.fit(X_adasyn_features, y_adasyn_target)

y_train_pred_adasyn = dt_regressor_adasyn.predict(X_adasyn_features)
y_test_pred_adasyn = dt_regressor_adasyn.predict(X_test)

rmse_train_adasyn = np.sqrt(mean_squared_error(y_adasyn_target, y_train_pred_adasyn))
r2_train_adasyn = r2_score(y_adasyn_target, y_train_pred_adasyn)

rmse_test_adasyn = np.sqrt(mean_squared_error(y_test, y_test_pred_adasyn))
r2_test_adasyn = r2_score(y_test, y_test_pred_adasyn)

print("\nModel Performance on ADASYN Train Data:")
print(f"RMSE: {rmse_train_adasyn}")
print(f"R2 Score: {r2_train_adasyn}")

print("\nADASYN Trained Model Performance on Test Data:")
print(f"RMSE: {rmse_test_adasyn}")
print(f"R2 Score: {r2_test_adasyn}")

y_train_pred_adasyn_category = pd.Series(y_train_pred_adasyn).apply(categorize_price)
y_test_pred_adasyn_category = pd.Series(y_test_pred_adasyn).apply(categorize_price)

accuracy_train_adasyn = accuracy_score(y_adasyn, y_train_pred_adasyn_category)
accuracy_test_adasyn = accuracy_score(y_test_category, y_test_pred_adasyn_category)

print("\nModel Accuracy on ADASYN Train Data:")
print(f"Accuracy: {accuracy_train_adasyn}")

print("\nADASYN Trained Model Accuracy on Test Data:")
print(f"Accuracy: {accuracy_test_adasyn}")
```

```python
# train random forest regressor on X_train and y_train

from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=200, random_state=42)
rf_regressor.fit(X_train, y_train)

y_train_pred_rf = rf_regressor.predict(X_train)
y_test_pred_rf = rf_regressor.predict(X_test)

rmse_train_rf = np.sqrt(mean_squared_error(y_train, y_train_pred_rf))
r2_train_rf = r2_score(y_train, y_train_pred_rf)

rmse_test_rf = np.sqrt(mean_squared_error(y_test, y_test_pred_rf))
r2_test_rf = r2_score(y_test, y_test_pred_rf)

print("Random Forest Model Performance on Train Data:")
print(f"RMSE: {rmse_train_rf}")
print(f"R2 Score: {r2_train_rf}")

print("\nRandom Forest Model Performance on Test Data:")
print(f"RMSE: {rmse_test_rf}")
print(f"R2 Score: {r2_test_rf}")

y_train_pred_rf_category = pd.Series(y_train_pred_rf).apply(categorize_price)
y_test_pred_rf_category = pd.Series(y_test_pred_rf).apply(categorize_price)

accuracy_train_rf = accuracy_score(y_train_category, y_train_pred_rf_category)
accuracy_test_rf = accuracy_score(y_test_category, y_test_pred_rf_category)

print("\nModel Accuracy on Random Forest Train Data:")
print(f"Accuracy: {accuracy_train_rf}")

print("\nRandom Forest Model Accuracy on Test Data:")
print(f"Accuracy: {accuracy_test_rf}")
```
